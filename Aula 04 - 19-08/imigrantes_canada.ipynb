```python
import pandas as pd #Importa a biblioteca Pandas
```

```python
df = pd.read_csv('imigrantes_canada.csv') #Importa a base de dados
```

```python
df.head()#Mostra as 4 primeiras linhas do dataframe
```

```python
df.info()#exibe as informações do dataframe
```

Tratamento de dados

```python
df.set_index(['País'],inplace=True) # seta o index com a coluna País
```

```python
anos = list(map(str,range(1980,2013))) #converte os anos para string armaenados em uma lista
```

```python
df.head()
```

```python
brasil = df.loc['Brasil',anos] #pega os dados somente para um pais
```

```python
brasil
```

```python
# cria um dicionario com os dados do brasil
brasil_dic={
    'ano':brasil.index.tolist(),
    'imigrantes':brasil.values.tolist(),
}
```

```python
df_brasil = pd.DataFrame(brasil_dic) #cria o dataframe
df_brasil # exibe os valores
```

```python
df_brasil.info()
```

```python
import matplotlib.pyplot as plt #importa a biblioteca matplot
```

```python
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])#plota o grafico
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)#ajuste de valores do eixo X
plt.yticks([500,1000,1500,2000,2500,3000],fontsize=14) # ajuste de valores do eixo Y
plt.title('Imigração do Brasil para o Canadá') # define o titulo
plt.xlabel('Anos',fontsize=14)#Titulo eixo X
plt.ylabel('Numeros de imigrantes', fontsize=14)# Titulo eixo Y
plt.show() # Exibe o grafico
```

```python
argentina = df.loc['Argentina', anos]
argentina
```

```python
argentina_dic={
    'ano':argentina.index.tolist(),
    'imigrantes':argentina.values.tolist(),
}
```

```python
df_argentina = pd.DataFrame(argentina_dic)
plt.plot(df_argentina['ano'],df_argentina['imigrantes'], label='Argentina')#plota o grafico
plt.plot(df_brasil['ano'],df_brasil['imigrantes'],label = 'Brasil')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)#ajuste de valores do eixo X
plt.yticks([500,1000,1500,2000,2500,3000],fontsize=14) # ajuste de valores do eixo Y
plt.title('Imigração da Argentina para o Canadá') # define o titulo
plt.xlabel('Anos',fontsize=14)#Titulo eixo X
plt.legend()
plt.ylabel('Numeros de imigrantes', fontsize=14)# Titulo eixo Y
plt.show() # Exibe o grafico
```

```python
df_comparação = df.loc[['Brasil','Argentina'], anos]#pega duas colunas Brasil e Argentina
```

```python
df_comparação = df_comparação.T # matriz trasposta
```

```python
df_comparação.head()
```

```python
plt.plot(df_comparação['Brasil'], label = 'Brasil',color='green')
plt.plot(df_comparação['Argentina'], label ='Argentina', color = 'blue')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.legend()
plt.show()
```

```python
df_comparação['Total'] = df_comparação['Brasil']+df_comparação['Argentina']
```

```python
plt.figure(figsize=(8,4))
plt.plot(df_comparação['Brasil'], label = 'Brasil',color='green')
plt.plot(df_comparação['Argentina'], label ='Argentina', color = 'blue')
plt.plot(df_comparação['Total'],label='Total', color = 'black')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.legend(loc='upper left')
plt.show()
```

```python
fig, ax =plt.subplots(figsize=(8,4)) # criando uma figura

ax.plot(df_brasil['ano'],df_brasil['imigrantes'],lw=3 , color ='green')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.grid(linestyle='--')
ax.set_title('Imigração do Brasil para o Canada \n 1980, a 2013', loc ='left')
ax.set_ylabel('Nº de imigrantes', fontsize=14)
ax.set_xlabel('Ano', fontsize = 14)

plt.show()
```

```python
# Criando a figura
fig, axs = plt.subplots(1, 2, figsize=(15,5))

# Subplot 1: Gráfico de linhas
axs[0].plot(df_brasil['ano'], df_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')

axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()
# Subplot 2: Boxplot
axs[1].boxplot(df_brasil['imigrantes'])
axs[1].set_title('Box-plot da imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid()
```
