```python
import pandas as pd #importa a biblioteca pandas
import matplotlib.pyplot as plt #importa a biblioteca matplotlib 
import numpy as np #importa a biblioteca numpy
```

```python
dados_x =np.linspace(0,10,10)#gera valores com a biblioteca numpy
dados_y = dados_x * dados_x #gera valores multiplicando um dado pelo o outro
```

```python
plt.figure(figsize=(10,5),facecolor='gray')#figsize -> define o tamanho do grafico
Ax =plt.axes() #cria os eixos do grafico
Ax.set_facecolor('white')#cria o fundo grafico
plt.plot(dados_x,dados_y,
         color='red', linewidth=2,#color -> cor da linha e linewidth -> espessura da linha
         marker='o', linestyle = 'dashed', markeredgecolor='black')#marker -> onde esta detacado o as informações no grafico / linestyle -> estilo da linha no grafico / -> markeregdecolor -> cor da borda do marker
plt.xlabel("Eixo X",fontsize=12)#adiciona o titulo do eixo x 
plt.ylabel("Eixo Y",fontsize=12)#adiciona o titulo do eixo y
plt.title("Exemplo matplotlib",fontsize=16)#define o titulo do grafico
```

```python
plt.rcParams.update({'font.size':15})#define o tamanho da fonte
plt.figure(figsize=(12,12))# define o tamanho da figura
linhas = 2#define o numeros de linhas 
colunas = 2#define o numero de colunas

#Grafico 01
plt.subplot(linhas, colunas,1)#define o numero de colunas e linhas do grafico
plt.plot(dados_x,'r--')
plt.title("Grafico 01")

#Grafico 02
plt.subplot(linhas, colunas,2)#define o numero de colunas e linhas do grafico
plt.plot(dados_y,'r--')
plt.title("Grafico 02")

#Grafico 03
plt.subplot(linhas, colunas,3)#define o numero de colunas e linhas do grafico
plt.plot(dados_x * dados_y)
plt.title("Grafico 03")

#Grafico 04
plt.subplot(linhas, colunas,4)#define o numero de colunas e linhas do grafico
plt.plot(dados_x / dados_y)
plt.title("Grafico 04")
plt.savefig("Exemplo subplot.jpg", dpi=300,transparent=True,
            bbox_inches="tight")#comando para salvar a imagem 
```

```python
#Grafico de barras
Figura = plt.figure()#Armazena na variavel Figura a a figura a se
Grafico = Figura.add_axes([0,0,0.7,0.7])# cria a grade do grafico
Rotulos = ['A','B','C','D','E','F','G','H','I','J']# cria os rotulos do grafico

Grafico.bar(Rotulos,dados_x, color='blue')
Grafico.bar(Rotulos,dados_y*0.1,color='red')
plt.title("Quantidade x Categoria")
plt.xlabel("Itens")
plt.ylabel("Quantidade")
```
