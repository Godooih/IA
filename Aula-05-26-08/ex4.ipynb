```python
import numpy as np # importa a biblioteca numpy
from sklearn.linear_model import LinearRegression # da biblioteca

import matplotlib.pyplot as plt
```

```python
# Dados de treinamento
renda = np.array([1200,5000,8000,10000,30000]).reshape(-1,1)#reshape serve para deixar a array na mesma linha
# Area do imovel
gastos = np.array([750,1500,4000,6000,18000])  # Valor do imovel 
#em Cº
# Criar e treinar o modelo de regressão linear
modelo = LinearRegression() # cria uma variavel chamada modelo
modelo.fit(renda, gastos) # cria o modelo com base nos dados de entrada
```

```python
renda_teste = np.array([5000,2000,1520,1800,1980,4750,15000,34560]).reshape(-1,1)

gastos_prevista = modelo.predict(renda_teste)
```

```python
for i in range(len(renda_teste)):
    print(f"Renda: R${renda_teste[i][0]},00 - Gasto previsto: R${gastos_prevista[i]:,.2f}")
plt.scatter(renda, gastos, color='blue', label='Dados de Treinamento')
plt.plot(renda_teste,gastos_prevista, color  = 'red',label='Dados de teste')
plt.xlabel('Renda')
plt.ylabel('Gastos')
plt.title('Relação entre Renda e Gastos')
plt.legend()
plt.grid(True)
plt.show()
```
