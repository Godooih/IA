```python
import numpy as np # importa a biblioteca numpy
from sklearn.linear_model import LinearRegression # da biblioteca

import matplotlib.pyplot as plt
```

```python
# Dados de treinamento
Consumo = np.array([50,140,30,75,123]).reshape(-1,1)#reshape serve para deixar a array na mesma linha
# Area do imovel
Demanda = np.array([100,200,80,100,160])  # Valor do imovel 
#em Cº
# Criar e treinar o modelo de regressão linear
modelo = LinearRegression() # cria uma variavel chamada modelo
modelo.fit(Consumo, Demanda) # cria o modelo com base nos dados de entrada
```

```python
Consumo_teste = np.array([20,36,56,80,79,520,314]).reshape(-1,1)

Demanda_prevista = modelo.predict(Consumo_teste)
```

```python
for i in range(len(Consumo_teste)):
    print(f"Consumo: {Consumo_teste[i][0]} KW - Demanda prevista:{Demanda_prevista[i]:,.2f}")
plt.scatter(Consumo, Demanda, color='blue', label='Dados de Treinamento')
plt.plot(Consumo_teste,Demanda_prevista, color  = 'red',label='Dados de teste')
plt.xlabel('Consumo (KW)')
plt.ylabel('Demanda')
plt.title('Relação entre Consumo e demanda')
plt.legend()
plt.grid(True)
plt.show()
```
