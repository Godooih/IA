```python
import numpy as np # importa a biblioteca numpy
from sklearn.linear_model import LinearRegression # da biblioteca

import matplotlib.pyplot as plt
```

```python
# Dados de treinamento
area = np.array([50,100,120,300]).reshape(-1,1)#reshape serve para deixar a array na mesma linha
# Area do imovel
valor = np.array([180.000,300.000,375.000,600.000])  # Valor do imovel 
#em Cº
# Criar e treinar o modelo de regressão linear
modelo = LinearRegression() # cria uma variavel chamada modelo
modelo.fit(area, valor) # cria o modelo com base nos dados de entrada
```

```python
area_teste = np.array([120,50,60,30,48,600,85,78,234,751,100]).reshape(-1,1)

valor_prevista = modelo.predict(area_teste)
```

```python
for i in range(len(area_teste)):
    print(f"Area: {area_teste[i][0]} m² - Valor previsto: R${valor_prevista[i]:,.3f},00")
plt.scatter(area, valor, color='blue', label='Dados de Treinamento')
plt.plot(area_teste,valor_prevista, color  = 'red',label='Dados de teste')
plt.xlabel('Area M²')
plt.ylabel('Valor(R$)')
plt.title('Relação entre Temperatura e Calor')
plt.legend()
plt.grid(True)
plt.show()
```
