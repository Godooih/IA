```python
import numpy as np # importa a biblioteca numpy
from sklearn.linear_model import LinearRegression # da biblioteca

import matplotlib.pyplot as plt

# sklearn importa o modelo de regressao linear
```

```python
# Dados de treinamento
calor = np.array([100,200,300,400,500,600,700]).reshape(-1,1)#reshape serve para deixar a array na mesma linha
# energia liberada 
temperatura = np.array([30,35,40,45,60,100,150])  # temperatura  
#em Cº
# Criar e treinar o modelo de regressão linear
modelo = LinearRegression() # cria uma variavel chamada modelo
modelo.fit(calor, temperatura) # cria o modelo com base nos dados de entrada
```

```python
calor_teste = np.array([150,250,350,450,550]).reshape(-1,1)

temperatura_prevista = modelo.predict(calor_teste)
```

```python
# Imprimir as previsões
for i in range(len(calor_teste)):
    print(f"Temperatura: {calor_teste[i][0]} joules - Temperatura previsto: {temperatura_prevista[i]} Cº")
plt.scatter(calor, temperatura, color='blue', label='Dados de Treinamento')
plt.plot(calor_teste,temperatura_prevista, color  = 'red',label='Dados de teste')
plt.xlabel('Calor (joules)')
plt.ylabel('Temperatura (Cº)')
plt.title('Relação entre Temperatura e Calor')
plt.legend()
plt.grid(True)
plt.show()
```
