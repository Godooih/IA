```python
import pandas as pd
```

```python
dicionario1 ={
    "Data":["2023-01-01","2023-01-02","2023-01-03"],
    "Produto":["Item A","Item B","Item A"],
    "Quantidade":['10','5','8'],
    "Valor":["100,00",'50,00','80,00']
}
Labels = ['','','']
```

```python
dados1 = pd.DataFrame(dicionario1, index=Labels)
```

```python
dados1.head()
```

```python
dicionario2 ={
    "Data":["2023-02-01","2023-02-02","2023-02-03"],
    "Produto":["Item C","Item B","Item D"],
    "Quantidade":['12','6','3'],
    "Valor":["120,00",'60,00','30,00']
}
Labels = ['','','']
```

```python
dados2 = pd.DataFrame(dicionario2, index= Labels)
```

```python
dados2.head()
```

```python
df_n = pd.concat([dados1,dados2]) # junta os dois Dataframe
```

```python
df_n.head(6)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(df_n['Produto'],df_n['Quantidade'],color = 'red') # define o tipo do grafico, informaes a serem analizadas e a cor do grafico
plt.xlabel("Nome do produto")
plt.xlabel("Quantidade")
plt.title('Produtos x Quantidade de vendas')
```
