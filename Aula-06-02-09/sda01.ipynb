```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# Dados da tabela
data = {
    'Velocidade de Operação': [10, 5, 8, 6, 12, 4],
    'Complexidade da Tarefa': ['Baixa', 'Alta', 'Média', 'Alta', 'Baixa', 'Alta'],
    'Manutenção Necessária': ['Baixa', 'Alta', 'Média', 'Alta', 'Baixa', 'Média'],
    'Classificação': ['Montagem', 'Teste', 'Montagem', 'Teste', 'Montagem', 'Teste']
}
```

```python
# Convertemos para DataFrame
df = pd.DataFrame(data)
```

```python
# Converter colunas categóricas para numéricas
df['Complexidade da Tarefa'] = df['Complexidade da Tarefa'].map({'Baixa': 0, 'Média': 1, 'Alta': 2})
df['Manutenção Necessária'] = df['Manutenção Necessária'].map({'Baixa': 0, 'Média': 1, 'Alta': 2})
df['Classificação'] = df['Classificação'].map({'Montagem': 0, 'Teste': 1})
```

```python
# Separando variáveis independentes (X) e variável dependente (y)
X = df[['Velocidade de Operação', 'Complexidade da Tarefa', 'Manutenção Necessária']]
y = df['Classificação']
```

```python
# Criar e treinar a árvore de decisão
clf = DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
# Visualizar a árvore de decisão
plt.figure(figsize=(10,8))
tree.plot_tree(clf, feature_names=['Velocidade de Operação', 'Complexidade da Tarefa', 'Manutenção Necessária'], class_names=['Montagem', 'Teste'], filled=True)
plt.show()
```

```python
# Testar o modelo com novos dados
novos_dados = [[7, 1, 1]]  # Exemplo: Velocidade 7, Complexidade Média, Manutenção Média
predicao = clf.predict(novos_dados)
print(f"Predição para os novos dados: {'Montagem' if predicao[0] == 0 else 'Teste'}")
```
