```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# Dados da tabela
data = {
    'Precisão na Montagem': ['Alta', 'Média', 'Alta', 'Média', 'Baixa', 'Baixa'],
    'Velocidade de Produção': ['Média', 'Baixa', 'Alta', 'Alta', 'Baixa', 'Média'],
    'Taxa de Retrabalho': ['Baixa', 'Alta', 'Baixa', 'Baixa', 'Alta', 'Alta'],
    'Classificação': ['Alta Qualidade', 'Baixa Qualidade', 'Alta Qualidade', 'Alta Qualidade', 'Baixa Qualidade', 'Baixa Qualidade']
}
```

```python
# Convertemos para DataFrame
df = pd.DataFrame(data)
```

```python

# Converter colunas categóricas para numéricas
df['Precisão na Montagem'] = df['Precisão na Montagem'].map({'Baixa': 0, 'Média': 1, 'Alta': 2})
df['Velocidade de Produção'] = df['Velocidade de Produção'].map({'Baixa': 0, 'Média': 1, 'Alta': 2})
df['Taxa de Retrabalho'] = df['Taxa de Retrabalho'].map({'Baixa': 0, 'Alta': 1})
df['Classificação'] = df['Classificação'].map({'Baixa Qualidade': 0, 'Alta Qualidade': 1})
```

```python
# Separando variáveis independentes (X) e variável dependente (y)
X = df[['Precisão na Montagem', 'Velocidade de Produção', 'Taxa de Retrabalho']]
y = df['Classificação']
```

```python
# Criar e treinar a árvore de decisão
clf = DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
# Visualizar a árvore de decisão
plt.figure(figsize=(12,8))
tree.plot_tree(clf, feature_names=['Precisão na Montagem', 'Velocidade de Produção', 'Taxa de Retrabalho'], class_names=['Baixa Qualidade', 'Alta Qualidade'], filled=True)
plt.show()
```

```python
# Testar o modelo com novos dados
novos_dados = [[2, 2, 1]]  # Exemplo: Precisão Alta, Velocidade Alta, Taxa de Retrabalho Alta
predicao = clf.predict(novos_dados)
print(f"Predição para os novos dados: {'Alta Qualidade' if predicao[0] == 1 else 'Baixa Qualidade'}")
```
